```python
%pylab inline
%run params.py

import matplotlib.pyplot as plt
import pandas as pd
import random
import os
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
from plotting_functions import *
```

```python
data = pd.read_csv(os.path.join(DATAPATH, "simulations_joint", "simulation_joint_results.txt"),
                   sep=" ", names=["logmu","coeff","est_logmu","est_coeff","est_logmu_stderr","est_coeff_stderr"])
```

```python
fig = plt.figure()
fig.set_size_inches((10, 3))
ax = fig.add_subplot(111)
ax.plot(range(data.shape[0]), data["logmu"], color="black", linewidth=1)
ax.scatter(range(data.shape[0]), data["est_logmu"], color="gray", s=40)
ax.set_xlim(left=-0.1, right=data.shape[0])
ax.set_ylim(bottom=-6.5, top=-1.0)
ax.set_xticks([])
ax.set_yticklabels(ax.get_yticks(), size=12)
ax.set_ylabel(r"Log10 $\mu_0$", size=15)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
fig.savefig(os.path.join(PLOTDIR, "SupplementalFigure13_Jointmodel_base.pdf"))

fig = plt.figure()
fig.set_size_inches((10, 3))
ax = fig.add_subplot(111)
ax.plot(range(data.shape[0]), data["coeff"], color="black", linewidth=1)
ax.scatter(range(data.shape[0]), data["est_coeff"], color="gray", s=40)
ax.set_xlim(left=-0.1, right=data.shape[0])
ax.set_ylim(bottom=-0.005, top=0.04);
ax.set_xticks([]);
ax.set_yticks(np.arange(0, 0.041, 0.01))
ax.set_yticklabels(ax.get_yticks(), size=12)
ax.set_ylabel(r"Coefficient $\theta$", size=15);
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom');
fig.savefig(os.path.join(PLOTDIR, "SupplementalFigure13_Jointmodel_lencoeff.pdf"))
```
